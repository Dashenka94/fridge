```python
# I need to create new column - PRODUCT,to extract products names from ingredients column without any measurement units


#1. I need to create function to Extract product names by splitting the cleaned ingredient list
def extract_product_names(ingredient):
    
    #2. I remove measurement units and other unnecessary words from column INGREDIENT so it could find key word from user request in future

    cleaned_ingredient = re.sub(r'\b(?:cup of|cup|<| > |(| )|devided|package|slices|diced|cooked|<|U|>|devided|large|big|small|spoon of|grams?|kilograms?|pounds?|ounces?|teaspoons?|tablespoons?|'
                                 r'\d+/\d+|\u00BC|\u00BD|\u00BE)\b', '', ingredient, flags=re.IGNORECASE)
    cleaned_ingredient = re.sub(r'\d', '', cleaned_ingredient)
    
    # I use list comprehension  
    return ', '.join(product.strip() for product in cleaned_ingredient.split('+') if product.strip())
    # product.strip() for product in cleaned_ingredient.split(',') if product.strip() - this part strips any whitespace from the beginning and end of each element 
    # cleaned_ingredient.split('+') - this part split string into a list of strings using the comma (+) so we could clean each substring from whitespaces
    # ', '.join() -  result after filtering and manipulation, is joined back together into a single string using ', ' as the separator


    #applying def to INGREDIENTS column and creating new column PRODUCT with separate products in it
all_recipes['product'] = all_recipes['ingredient'].apply(extract_product_names)
```

```python
## 1 algorithm of matching

all_recipes['ingredient_match'] = all_recipes['product'].str.lower().apply(lambda x: all(ingredient in x for ingredient in available_ingredients))

# Filter recipes based on ingredient matching
matching_rows = all_recipes[all_recipes['ingredient_match']]

# Display recommended recipes
print("Recommended Recipes:")
matching_rows[['name', 'ingredient']]
```

```python
# to calculate number of products at users fridge

number_of_products_in_fridge = sum(isinstance(item, str) for item in available_ingredients)
```

```python
selected_recipe_ids = matching_rows['name']

# Subset the main dataset to include only the selected rows
selected_recipes = all_recipes[all_recipes['name'].isin(selected_recipe_ids)]

# Add a new column to the subset
selected_recipes['products_to_buy'] = selected_recipes['num_of_ingredients'] - number_of_products_in_fridge

sorted_recipes = selected_recipes.sort_values(by='products_to_buy', ascending=True)
```
